```python
import pandas as pd
import numpy as np
from scipy.stats import rankdata
import os

BASE_DIR = "../outputs"   # ðŸ‘ˆ ì¤‘ìš”

files = {
    "fixed": "10_FIXED_submission_CV0.740810.csv",
    "ag": "03_submission_autogluon(0206).csv",
    "oof": "05_submission_w0.72_CV0.740466.csv",
    "stack": "02_submission_stacking_v1(0206).csv",
    "rf": "03.3_seed123_CV0.744878.csv",
    "cb_only": "05_sub_cb_only.csv"
}

ranks = {}
for k, fname in files.items():
    path = os.path.join(BASE_DIR, fname)
    df = pd.read_csv(path)
    ranks[k] = rankdata(df["probability"].values)

keys = list(ranks.keys())
corr = np.zeros((len(keys), len(keys)))

for i, a in enumerate(keys):
    for j, b in enumerate(keys):
        corr[i, j] = np.corrcoef(ranks[a], ranks[b])[0, 1]

corr_df = pd.DataFrame(corr, index=keys, columns=keys)
print(corr_df.round(4))
```
