### 2. 특정 시술 유형

```python
df["특정 시술 유형"].value_counts()
```

#### 대분류로 구분해두고 target 비교

```python
def major_procedure(x):
    if pd.isna(x):
        return "Unknown"
    if "IUI" in x:
        return "IUI"
    if "DI" in x:
        return "DI"
    if "ICSI" in x:
        return "ICSI"
    if "IVF" in x:
        return "IVF"
    if "FER" in x:
        return "FER"
    if "GIFT" in x:
        return "GIFT"
    return "Other"

df["시술_대분류"] = df["특정 시술 유형"].apply(major_procedure)

rate = (
    df.groupby("시술_대분류")["임신 성공 여부"]
      .agg(["count", "mean"])
      .rename(columns={"mean": "success_rate"})
      .sort_values("success_rate", ascending=False)
)

rate
```

```python
# FER, GIFT, DI, Unknown sample수가 너무 적음 -> Other로 통합
rare = ["FER", "GIFT", "Unknown", "DI"]
df["시술_대분류"] = df["시술_대분류"].replace(rare, "Other")
```

```python
# 합치고 다시 확인
rate = (
    df.groupby("시술_대분류")["임신 성공 여부"]
      .agg(["count", "mean"])
      .rename(columns={"mean": "success_rate"})
      .sort_values("success_rate", ascending=False)
)

rate
```

```python
# 통계적으로 차이가 있는지 (p<0.05)
ct = pd.crosstab(df["시술_대분류"], df["임신 성공 여부"])
chi2, p, _, _ = stats.chi2_contingency(ct)
print(f"chi2 p-value: {p:.3e}")
```

```python
# 세부적으로 (Blastocyst | AH)
df["BLASTOCYST_포함"] = df["특정 시술 유형"].str.contains("BLASTOCYST", na=False).astype(int)
df["AH_포함"]         = df["특정 시술 유형"].str.contains("AH", na=False).astype(int)
```

```python
# Blastocyst 여부에 따른 임신 성공 여부 차이 
df.groupby("BLASTOCYST_포함")["임신 성공 여부"] \
  .agg(["count", "mean"]) \
  .rename(columns={"mean": "success_rate"})
```

```python
# AH 여부에 따른 임신 성공 여부 차이
df.groupby("AH_포함")["임신 성공 여부"] \
  .agg(["count", "mean"]) \
  .rename(columns={"mean": "success_rate"})
```
